```python
import pygad
import numpy as np
import utils
```

```python
hidden_layers = 5
activation_functions = ['tanh', 'tanh', 'tanh','tanh','tanh']
initial_pop = [[50, 50, 50, 30, 30],
               [200, 100, 200, 40, 40]]
```

```python
def fitness_func(pygadClass, solution, solution_idx):
    if np.all(solution > 0):
        output = utils.evaluateANN(hidden_layers, solution, activation_functions)
        print(output)
        fitness = 1 / output
        return fitness
    else:
        return 0
```

```python
ga_instance = pygad.GA(num_generations=50,
                       sol_per_pop=10,
                       #initial_population=initial_pop,
                       num_genes=hidden_layers,
                       num_parents_mating=2,
                       gene_type=int,
                       fitness_func=fitness_func,
                       init_range_low=25,
                       init_range_high=200,
                       mutation_type="random",
                       mutation_probability=0.6,
                       
                       crossover_type="uniform",
                       crossover_probability=0.5,
                       
                       #save_solutions=True,
                       save_best_solutions=True)

ga_instance.run()
```

```python
print(ga_instance.initial_population)
print("-----------------------------")
print(ga_instance.population)
ga_instance.plot_fitness()
```
