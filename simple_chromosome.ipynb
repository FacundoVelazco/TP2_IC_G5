```python
import pygad
import numpy as np
import utils
```

```python
hidden_layers = 3
activation_functions_5 = ['tanh', 'tanh', 'tanh','tanh','tanh']
activation_functions_3 = ['tanh', 'tanh', 'tanh']

initial_pop_5 = [[50, 50, 50, 30, 30],
               [200, 100, 200, 40, 40],
               [200, 100, 200, 40, 40],
               [200, 100, 200, 40, 40],
               [200, 100, 200, 40, 40],
               [200, 100, 200, 40, 40],
               [200, 100, 200, 40, 40],
               [200, 100, 200, 40, 40],
               [200, 100, 200, 40, 40],
               [200, 100, 200, 40, 40]]

initial_pop_3 = [[25, 45, 25],
               [25,50,25],
               [37, 57, 37],
               [23, 30, 23]]
```

```python
def fitness_func(pygadClass, solution, solution_idx):
    if np.all(solution > 0):
        output = utils.evaluateANN(hidden_layers, solution, activation_functions_3)
        print("My config: ",solution," ; MSE:",output)
        fitness = 1 / output * 10000
        return fitness
    else:
        return 0
```

```python
ga_instance = pygad.GA(num_generations=10,
                       sol_per_pop=10,
                       #initial_population=initial_pop_3,
                       num_genes=hidden_layers,
                       num_parents_mating=2,
                       gene_type=int,
                       fitness_func=fitness_func,
                       init_range_low=25,
                       init_range_high=200,
                       
                       mutation_type="random",
                       random_mutation_min_val=1,
                       random_mutation_max_val=50,
                       mutation_percent_genes=0.666,
                       mutation_probability=0.6,
                       
                       parent_selection_type="rank",
                       
                       crossover_type="uniform",
                       crossover_probability=0.5,
                       
                       keep_elitism=5,
                       
                       allow_duplicate_genes=True,
                       
                       suppress_warnings=True,
                       save_solutions=True,
                       save_best_solutions=True)

ga_instance.run()
```
